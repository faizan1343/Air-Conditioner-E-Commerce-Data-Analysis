```python
import numpy as np
df['actual_price'] = df['actual_price'].astype(str)
df['discount_price'] = df['discount_price'].astype(str)

# Remove '₹' symbol and commas, and convert 'actual_price' and 'discount_price' to numeric
df['actual_price'] = pd.to_numeric(df['actual_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')
df['discount_price'] = pd.to_numeric(df['discount_price'].str.replace('₹', '').str.replace(',', ''), errors='coerce')

# Exclude rows with NaN values in either 'actual_price' or 'discount_price'
df = df.dropna(subset=['actual_price', 'discount_price'])

# Calculate the discounted price and store it in the 'Discounted_Price' column
df['Discounted_Price'] = df['actual_price'] - df['discount_price']

# Calculate the discount percentage and store it in the 'Discount_Percentage' column
df['Discount_Percentage'] = ((df['Discounted_Price'] / df['actual_price']) * 100).round(2)

# Display the updated DataFrame
df
```

```python
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')

# Calculate the average rating
average_rating = df['ratings'].mean()

# Display the average rating

average_discount_price = df['discount_price'].mean()
average_actual_price = df['actual_price'].mean()
average_discounted_price = df['Discounted_Price'].mean()
average_discount_percentage = df['Discount_Percentage'].mean()

# Display the averages
print(f"Average Rating: {average_rating:.2f}")
print(f"Average discount_price: ₹{average_discount_price:.2f}")
print(f"Average actual_price: ₹{average_actual_price:.2f}")
print(f"Average Discounted_Price: ₹{average_discounted_price:.2f}")
print(f"Average Discount_Percentage: {average_discount_percentage:.2f}%")
```
